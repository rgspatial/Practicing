```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load Fifa Data
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

## Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa['Overall'], bins=bins, color='#b82845')

plt.xticks(bins)

plt.ylabel("Number of Players")
plt.xlabel("Skill Level")
plt.title("Distribution of Players Skill Level - FIFA")

plt.show()
```

## Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#eb7434', '#3723b8']

plt.pie([left, right], labels=labels, colors=colors, autopct="%.2f%%")

plt.title('Foot Preference FIFA')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

plt.figure(figsize=(5,3), dpi=200)

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight <150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4, .2, 0, 0, .4)

plt.title('Weight FIFA Players')

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=.7, explode=explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=[5,7])

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
cork = fifa.loc[fifa.Club == 'Cork City']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'Cork City']

boxes = plt.boxplot([barcelona, madrid, cork], labels = labels, patch_artist=True, medianprops={'linewidth':2})
for box in boxes['boxes']:
    box.set(color='#2c226b', linewidth=2)
    box.set(facecolor='#1f6954')

plt.title('Professional Futball Team Comparison')
plt.ylabel('FIFA Overall Rating')


plt.show()
```
