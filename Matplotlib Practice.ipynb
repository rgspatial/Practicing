```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic Graph

```python
x =[0,1,2,3,4]
y=[0,2,4,6,8]

#resize graph
plt.figure(figsize=(4,2), dpi=200)

#Line 1
plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', markersize=15)

#line number two
#select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

#plot part of graph as line and part as dashed
plt.plot(x2[:6], x2[:6]**2, label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'b--')

#add a title (specify font parameters with fontdict)
plt.title("Graph Practice", fontdict={'fontname': 'Arial', 'fontsize': 15})

#x and y labels
plt.xlabel("X Axis", fontdict={'fontname': 'Arial'})
plt.ylabel("Y Axis", fontdict={'fontname': 'Arial'})

#x and y tickmarks
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

#add legend
plt.legend()

#save image
plt.savefig('practice_graph.png', dpi=300)

#show plot
plt.show()
```

Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)
bars[0].set_hatch("/")
bars[1].set_hatch("x")
bars[2].set_hatch("|")

plt.figure(figsize=(6,4))

plt.show()
```
